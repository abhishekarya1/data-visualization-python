```python
from matplotlib import pyplot as plt
```

```python
tags = ['A', 'B', 'C', 'D', 'E']
slices = [100, 200, 500, 400, 200]
```

```python
plt.style.use('fivethirtyeight')
plt.title('Drawing Pie Chart')

plt.pie(slices, labels = tags)
```

```python
explode_arr = [0, 0, 0, 0.1, 0]
plt.pie(slices, labels = tags, 
        wedgeprops = {'edgecolor': 'black'}, 
        explode = explode_arr, 
        shadow = True,
        startangle = 90,
        autopct = '%1.1f%%'
       )
```
