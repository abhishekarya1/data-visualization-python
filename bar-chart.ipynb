```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x_vals = ['A', 'B', 'C', 'D', 'E']
y_vals = [100, 200, 500, 400, 200]
```

```python
plt.title('Drawing Bar Chart')
plt.xlabel('Name')
plt.ylabel('Price (USD)')

plt.bar(x_vals, y_vals)
```

### Horizontal Bar Chart

```python
plt.barh(x_vals, y_vals)
```

### Side-by-Side Bars
For placing bars side-by-side, offset by `width` parameter using numpy arrray, need to customize ticks for correct x-axis labels.

```python
width = 0.25
y2_vals = [50, 500, 150, 300, 200]
y3_vals = [100, 300, 400, 250, 50]

x_pos = np.arange(len(x_vals))

plt.xticks(ticks = x_pos, labels = x_vals)

# use numpy object as x-axis
plt.bar(x_pos - width, y_vals, width = width)
plt.bar(x_pos, y2_vals, width = width)
plt.bar(x_pos + width, y3_vals, width = width)
```

### Stacked Bars
Stack bars using `bottom` parameter 

```python
plt.bar(x_pos, y_vals)
plt.bar(x_pos, y2_vals, bottom = y_vals)
```
