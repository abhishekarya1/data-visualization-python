Stack Plots are useful to compare contribution to tha total (data points on Y-axis) for a given point on X-axis.

```python
from matplotlib import pyplot as plt
```

```python
x_vals = [1, 2, 3, 4, 5]
y1_vals = [100, 200, 500, 400, 50]
y2_vals = [10, 200, 50, 70, 30]
y3_vals = [100, 20, 300, 60, 20]
```

```python
plt.title('Drawing Stack Plots')
colors = ['red', 'green', 'yellow']
labels = ['Tomato', 'Pear', 'Mango']

plt.stackplot(x_vals, y1_vals, y2_vals, y3_vals, colors = colors, labels = labels)

plt.legend(loc = (0.02, 0.80))
```
