```python
from matplotlib import pyplot as plt
```

```python
x_vals = [1, 2, 3, 4, 5]
y_vals = [100, 200, 500, 400, 200]
y2_vals = [100, 50, 1000, 300, 10]
```

```python
# plt.xkcd()
plt.style.use('fivethirtyeight')

plt.title('Drawing Simple Plots')
plt.xlabel('Length (metres)')
plt.ylabel('Price (USD)')

plt.plot(x_vals, y_vals, color='#3b3b3b', label='foo')
plt.plot(x_vals, y2_vals, color='g', linestyle='--', label='bar', linewidth=2, marker='o')

plt.legend()
plt.grid(True)

plt.tight_layout()
```
